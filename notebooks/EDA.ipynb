```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
multithreading = pd.read_csv('../results-multithreading.csv')
multiprocessing = pd.read_csv('../results-multiprocessing.csv')
```

```python
plt.plot(multithreading['File Count'], multithreading['Avg. Time'], label='Multithreading')
plt.plot(multiprocessing['File Count'], multiprocessing['Avg. Time'], label='Multiprocessing')
plt.legend()
plt.xlabel('File Count')
plt.ylabel('Avg. Time (s)')
```
